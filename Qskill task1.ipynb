```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/adika/Downloads/titanic/train.csv")

print(df.shape)
print(df.isnull().sum())
print(df.describe())
```

```python
print("Average Age:", df["Age"].mean())
print("Average Fare:", df["Fare"].mean())
print("Survival Rate:", df["Survived"].mean()*100)
```

Survival Count

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution")
plt.show()
```

OBSERVATION
1) More passengers died than survived.
2) Around 38% survived.

SURVIVAL BY GENDER


```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

Observation
1) Females had a much higher survival rate.
2) Males were more likely to die.

Survival by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

Observation
1) First-class passengers had the highest survival rate.
2) Third-class passengers had the lowest survival rate.

Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

Observation
1. Most passengers were between 20 and 40 years old.
2. Fewer elderly passengers were present.

Fare Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title("Fare Distribution")
plt.show()
```

Observation
1. Most passengers paid low fares.
2. A few passengers paid very high fares (outliers).

Age vs Fare Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

Observation
1. No strong relationship between age and fare.
2. High fares were paid by passengers of different ages.

Correlation Heatmap

```python
plt.figure(figsize=(10,8))

numeric_df = df.select_dtypes(include='number')

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

Observation
1. Fare positively correlates with survival.
2. Pclass negatively correlates with survival.
3. Age has weak correlation with survival.

Survival by Embarked Port

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival by Embarkation Port")
plt.show()
```

Observation
1. Passengers embarking from Cherbourg (C) had the highest survival rate.

Family Size Analysis

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

plt.figure(figsize=(8,5))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Family Size vs Survival")
plt.show()
```

Observation
1. Small families had better survival chances.
2. Very large families had lower survival rates.

Box Plot of Fare by Class

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

Observation
1. First-class passengers paid significantly higher fares.
2. Third-class fares were generally low.
